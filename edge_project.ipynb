```python
import csv

# File to store student data
file_name = "student_data.csv"

# Ensure the CSV file has a header if it doesn't exist
try:
    with open(file_name, mode="x", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Student Name", "Student ID", "Subject", "Score", "Attendance Percentage", "Assignment Completion"])
except FileExistsError:
    pass  # File already exists, no need to create a new one

# Function to add a new entry
def add_entry():
    with open(file_name, mode="a", newline="") as file:
        writer = csv.writer(file)
        print("\nEnter student data (leave 'Student Name' blank to stop):")
        while True:
            student_name = input("Student Name: ")
            if not student_name:  # Stop when no name is entered
                break
            student_id = input("Student ID: ")
            subject = input("Subject: ")
            score = input("Score: ")
            attendance_percentage = input("Attendance Percentage: ")
            assignment_completion = input("Assignment Completion (Yes/No): ")
            writer.writerow([student_name, student_id, subject, score, attendance_percentage, assignment_completion])
        print("Entries added successfully!")

# Function to search for a student
def search_student():
    search_term = input("\nEnter Student Name or ID to search: ").strip()
    with open(file_name, mode="r") as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row
        found = False
        print("\nSearch Results:")
        for row in reader:
            if search_term in row:  # Search by name or ID
                print(dict(zip(header, row)))  # Print matching row as a dictionary
                found = True
        if not found:
            print("No student found with the given name or ID.")

# Function to update a student's score or attendance
def update_score_attendance():
    student_id = input("\nEnter Student ID to update: ").strip()
    updated_rows = []
    found = False

    # Read and update data
    with open(file_name, mode="r") as file:
        reader = csv.reader(file)
        header = next(reader)
        updated_rows.append(header)  # Keep the header

        for row in reader:
            if row[1] == student_id:  # Check Student ID
                print(f"\nCurrent Record: {dict(zip(header, row))}")
                new_score = input("Enter new score (leave blank to keep current): ").strip()
                new_attendance = input("Enter new attendance percentage (leave blank to keep current): ").strip()
                if new_score:
                    row[3] = new_score
                if new_attendance:
                    row[4] = new_attendance
                found = True
            updated_rows.append(row)

    if found:
        # Write updated data back to file
        with open(file_name, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerows(updated_rows)
        print("Student record updated successfully!")
    else:
        print("No student found with the given ID.")

# Main menu
while True:
    print("\nMenu:")
    print("1. Add new entry")
    print("2. Search for a student")
    print("3. Update student's score or attendance")
    print("4. Exit")

    choice = input("Choose an option from 1 to 4: ").strip()

    if choice == "1":
        add_entry()
    elif choice == "2":
        search_student()
    elif choice == "3":
        update_score_attendance()
    elif choice == "4":
        print("Exiting the program...")
        break
    else:
        print("Invalid choice, please select a number from 1 to 4.")
```
