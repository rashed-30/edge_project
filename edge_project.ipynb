```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive

# Step 1: Mount Google Drive and Load Data
drive.mount('/content/drive')
path = "/content/drive/My Drive/Edge/student_data.csv"
df = pd.read_csv(path)
def categorize_attendance(value):
    if value < 50:
        return "<50%"
    elif 50 <= value <= 75:
        return "50-75%"
    else:
        return ">75%"

df["Attendance Category"] = df["Attendance Percentage"].apply(categorize_attendance)

# Count categories
category_counts = df["Attendance Category"].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    category_counts,
    labels=category_counts.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=["#66c2a5", "#fc8d62", "#8da0cb"],
    wedgeprops={"edgecolor": "black", "linewidth": 1}
)
plt.title("Attendance Percentage Categories", fontsize=14)
plt.show()

path2 = "/content/drive/My Drive/Edge/marks.csv"
df2 = pd.read_csv(path2)
selected_student = "Rashed"  # Replace with the desired student's name
subjects = ["Bangla", "English", "Math"]
performance = df2[df2["Student Name"] == selected_student][subjects].values.flatten()

plt.figure(figsize=(10, 6))
plt.plot(subjects, performance, marker="o", linestyle="-", color="g")
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Marks", fontsize=12)
plt.title(f"Performance Trends for {selected_student}", fontsize=14)
plt.ylim(0, 100)
plt.grid(True, linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

```

```python

import csv
from google.colab import drive
drive.mount('/content/drive')

# File to store student data
file_path = '/content/drive/My Drive/Edge/student_data.csv'

# Ensure the CSV file has a header if it doesn't exist
try:
    with open(file_path, mode="x", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["Student Name", "Student ID", "Subject", "Score", "Attendance Percentage", "Assignment Completion"])
except FileExistsError:
    pass  # File already exists, no need to create a new one

# Function to add a new entry
def add_entry():
    with open(file_path, mode="a", newline="") as file:
        writer = csv.writer(file)
        print("\nEnter student data (leave 'Student Name' blank to stop):")
        while True:
            student_name = input("Student Name: ")
            if not student_name:  # Stop when no name is entered
                break
            student_id = input("Student ID: ")
            subject = input("Subject: ")
            score = input("Score: ")
            attendance_percentage = input("Attendance Percentage: ")
            assignment_completion = input("Assignment Completion (Yes/No): ")
            writer.writerow([student_name, student_id, subject, score, attendance_percentage, assignment_completion])
        print("Entries added successfully!")

# Function to search for a student
def search_student():
    search_term = input("\nEnter Student Name or ID to search: ").strip()
    with open(file_path, mode="r") as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row
        found = False
        print("\nSearch Results:")
        for row in reader:
            if search_term in row:  # Search by name or ID
                print(dict(zip(header, row)))  # Print matching row as a dictionary
                found = True
        if not found:
            print("No student found with the given name or ID.")

# Function to update a student's score or attendance
def update_score_attendance():
    student_id = input("\nEnter Student ID to update: ").strip()
    updated_rows = []
    found = False

    # Read and update data
    with open(file_path, mode="r") as file:
        reader = csv.reader(file)
        header = next(reader)
        updated_rows.append(header)  # Keep the header

        for row in reader:
            if row[1] == student_id:  # Check Student ID
                print(f"\nCurrent Record: {dict(zip(header, row))}")
                new_score = input("Enter New Score (leave blank to keep current): ").strip()
                new_attendance = input("Enter New Attendance Percentage (leave blank to keep current): ").strip()
                new_assignment = input("Enter Assignment Status (leave blank to keep current): ").strip()
                if new_score:
                    row[3] = new_score
                if new_attendance:
                    row[4] = new_attendance
                if new_assignment:
                    row[5] = new_assignment
                found = True
            updated_rows.append(row)

    if found:
        # Write updated data back to file
        with open(file_path, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerows(updated_rows)
        print("Student record updated successfully!")
    else:
        print("No student found with the given ID.")


# Main menu
while True:
    print("\nMenu:")
    print("1. Add new entry")
    print("2. Search for a student")
    print("3. Update student's score or attendance")
    print("4. Exit")

    choice = input("Choose an option from 1 to 4: ").strip()

    if choice == "1":
        add_entry()
    elif choice == "2":
        search_student()
    elif choice == "3":
        update_score_attendance()
    elif choice == "4":
        print("Exiting the program...")
        break
    else:
        print("Invalid choice, please select a number from 1 to 4.")
```
